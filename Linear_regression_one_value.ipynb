**Step 1: Generate Sample Dataset**

```python
import numpy as np
import pandas as pd

# Set seed for reproducibility
np.random.seed(42)

# Generate data
area = np.random.randint(500, 4000, 100)  # area in sq. ft.
price = area * 50 + np.random.randint(10000, 50000, 100)  # linear relation with noise

# Create DataFrame
df = pd.DataFrame({'Area (sqft)': area, 'Price (₹)': price})
df.to_csv("house_data.csv", index=False)
```

**Step 2: Load and Preprocess Data**

```python
df = pd.read_csv("house_data.csv")

# Check for missing values
print(df.isnull().sum())

# View basic stats
print(df.describe())
#print(df.info())
#print(df.shape)
#print(df.columns)
#print(df.head())
```

```python
print(df.isnull().sum())
```

**Step 3: Exploratory Data Analysis (EDA)**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot
plt.scatter(df['Area (sqft)'], df['Price (₹)'], color='blue')
plt.title('Area vs Price')
plt.xlabel('Area (sqft)')
plt.ylabel('Price ($)')
plt.grid(True)
plt.show()

# Correlation heatmap
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

**Step 4: Build the Linear Regression Model**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
#from sklearn.model_selection import train_test_split
# Features and labels
X = df[['Area (sqft)']]
y = df['Price (₹)']

# Split into train-test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Train model
model = LinearRegression()
model.fit(X_train, y_train)

# Get slope and intercept
print("Coefficient:", model.coef_)
print("Intercept:", model.intercept_)
```

**Step 5: Model Evaluation**

```python
from sklearn.metrics import mean_squared_error, r2_score

# Predict on test data
y_pred = model.predict(X_test)

# Evaluate
print("MSE:", mean_squared_error(y_test, y_pred))
#print("R² Score:", r2_score(y_test, y_pred))
```

**Step 6: Prediction and Visualization**

```python
# Predict a new value
new_area = [[2500]]
predicted_price = model.predict(new_area)
print(f"Predicted price for 2500 sqft: ₹ {predicted_price[0]:,.2f}")

# Plot regression line
plt.scatter(X, y, color='blue')
plt.plot(X, model.predict(X), color='red')  # regression line
plt.title("Regression Line: Area vs Price")
plt.xlabel("Area (sqft)")
plt.ylabel("Price (₹)")
plt.grid(True)
plt.show()
```
